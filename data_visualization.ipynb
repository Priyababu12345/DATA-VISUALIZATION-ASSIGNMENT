```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility

# Generate 1000 samples of data
n_samples = 1000

# Create random features
x1 = np.random.normal(loc=0, scale=1, size=n_samples)
x2 = 0.8 * x1 + np.random.normal(loc=0, scale=0.5, size=n_samples)  # x2 is correlated with x1
x3 = 0.5 * x1 + 0.5 * x2 + np.random.normal(loc=0, scale=0.5, size=n_samples)  # x3 is correlated with x1 and x2
x4 = np.random.normal(loc=0, scale=1, size=n_samples)  # x4 is independent

# Create a DataFrame with these features
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Step 2: Calculate the correlation matrix
correlation_matrix = data.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=14)

# Display the plot
plt.show()
```

```python
# Example Interpretation:
# Diagonal: All diagonal elements in the correlation matrix are 1 because each feature is perfectly correlated with itself.
# Off-diagonal: For instance, the correlation between Feature 1 and Feature 2 might be 0.8, which indicates a strong positive correlation. Similarly, Feature 1 and Feature 3 might have a correlation of 0.5.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate the dataset
np.random.seed(25)

data = {
    'country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'population': np.random.randint(100, 1000, 5) * 1e6,
    'GDP': np.random.randint(500, 2000, 5) * 1e9
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Step 2: Create a bubble chart
plt.figure(figsize=(10, 6))

# Bubble chart - x: GDP, y: population, size: population (scaled for better visualization)
plt.scatter(df['GDP'], df['population'],
            s=df['population'] / 1e6,
            alpha=0.6,
            c=df['population'],
            cmap='viridis',
            edgecolors="w",
            linewidth=1)

# Step 3: Add labels and titles
for i in range(len(df)):
    plt.text(df['GDP'][i], df['population'][i], df['country'][i], ha='center', va='center', fontsize=10, color='black')

plt.title('Bubble Chart: GDP vs Population by Country', fontsize=14)
plt.xlabel('GDP (in billions USD)', fontsize=12)
plt.ylabel('Population (in millions)', fontsize=12)

# Display the plot
plt.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare data
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Create a Bokeh plot
plot = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
plot.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Configure plot legend and display it
plot.legend.location = "top_left"
output_notebook()
show(plot)
```
